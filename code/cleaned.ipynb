# Course Popularity & Engagement

<H3>1. Understanding Market Demand <br>
Why? - This helps identify popular subjects and what learners are most interested in.<br>
What are the most popular subjects based on the number of subscribers?</H3>

Why This Visualization?

Understanding which subject has the highest number of subscribers helps course creators and educational platforms focus on in-demand topics.

It also gives insights into learner preferences and industry trends.

```python
import matplotlib.ticker as mticker
```

```python
graph_width = 7.5 / 2.54   
graph_height = 5.5 / 2.54
```

```python
plt.figure(figsize=(graph_width, graph_height))
subject_subscribers = data.groupby("subject")["num_subscribers"].sum().reset_index()
sns.barplot(data=subject_subscribers,x='subject',y='num_subscribers',ci=None,color=udemy_purple)
plt.xlabel("Subject", fontsize=12, color=udemy_black)
plt.ylabel("Total Subscribers", fontsize=12, color=udemy_black)
plt.xticks(rotation=25, fontsize=8,color=udemy_black)
plt.gca().get_yaxis().set_major_formatter(mticker.FuncFormatter(lambda x,  _: f"{x/1e6:.1f}M"))
plt.savefig("performance_evaluation.png", dpi=300, bbox_inches='tight')
plt.show()
print(subject_subscribers)
```

Web Development has the most subscribers, showing it's the most popular subject. Other subjects like Business Finance and Graphic Design have fewer but steady learners.

<h3>2. Course Level Insights<br>
Why? - To see if beginner, intermediate, or advanced courses are more preferred.<br>
Do advanced-level courses get more subscribers than beginner or intermediate courses?</h3>

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=data,x='level',y='num_subscribers')
plt.title("Subscribers Distribution Across Course Levels")
plt.show()
```

Beginner Level and All Levels courses have the highest number of outliers, indicating some courses attract a significantly larger number of subscribers.

Expert Level courses have the least number of subscribers, with very few high-subscriber outliers.

Intermediate Level courses also have a few high-subscriber outliers but generally attract fewer subscribers than Beginner and All Levels courses.

The median number of subscribers is low across all levels, but a few courses dominate in terms of popularity.

Conclusion: Beginner and All Levels courses are the most popular, attracting the highest number of subscribers, while Expert Level courses have the least engagement.

<h3>3. Impact of Reviews on Engagement<br>
Question: Is there a relationship between the number of reviews and the number of subscribers?<br>
Why? To see if courses with more subscribers also receive more reviews, indicating engagement.<br>
Graph: Scatterplot (sns.scatterplot) – Shows the correlation between reviews and subscribers.</h3>

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=data,x="num_reviews",y="num_subscribers",alpha=0.5)
plt.title("Subscribers vs. Reviews")
plt.show()
```

Explanation:
This scatter plot visualizes the relationship between the number of reviews and the number of subscribers. Each dot represents a course, with its position determined by the number of reviews (x-axis) and the number of subscribers (y-axis).

The general trend suggests that courses with more subscribers tend to have more reviews, though there are some outliers. The density of points in the lower left indicates that most courses have relatively few subscribers and reviews.

Interpretation (Short Version):
There is a positive correlation between the number of subscribers and the number of reviews, meaning popular courses tend to receive more reviews.

# Pricing & Revenue Insights

<h3>4. Free vs Paid Courses<br>
Question: Do free courses attract more subscribers than paid ones?<br>
Why? To analyze whether pricing affects enrollments.<br>
Graph: Bar Chart (sns.barplot) – Compares the total number of subscribers for free vs. paid courses.
</h3>

```python
plt.figure(figsize=(10,8))
sns.barplot(data=data,x='is_paid',y='num_subscribers',ci=None,palette=['#A435F0'])
plt.title("Subscribers in Free vs. Paid Courses",color='#2D2F31',fontweight='bold',fontsize=14)
plt.show()
```

Interpretation:
Free courses have significantly more total subscribers compared to paid courses.

This suggests that users are more likely to enroll in free courses rather than paying for them.

However, this does not indicate quality or revenue—paid courses might have fewer students but could generate more income.

<h3>5. Price vs. Subscribers<br>
Question: Does a higher price mean fewer subscribers?<br>
Why? To see if high prices discourage learners or if premium courses perform well.<br>
Graph: Scatterplot (sns.scatterplot) – Helps visualize the trend between price and subscribers.</h3>

```python
plt.figure(figsize=(7,4))
sns.scatterplot(data=data,x="price",y="num_subscribers",alpha=0.6,color=udemy_purple)
plt.xlabel("Price ($)", fontsize=12, color=udemy_black)
plt.ylabel("Subscribers", fontsize=12, color=udemy_black)
plt.savefig("Price_vs_Enrollment.png", dpi=300, bbox_inches='tight')
plt.show()
```

Free courses (price = 0) have the highest number of subscribers, showing that more people enroll in free courses.

As the price increases, subscriber count generally decreases, indicating that fewer people enroll in expensive courses.

Some high-priced courses still have a considerable number of subscribers, suggesting demand for premium content.

Overall, price negatively correlates with subscribers, but there are exceptions where high-priced courses attract many learners.

# Content & Course Structure

<h3>7. Price vs. Content Depth<br>
Question: Does a higher price mean more lectures?<br>
Why? To check if expensive courses justify their price by offering more lectures.<br>
Graph: Scatterplot (sns.scatterplot) – Reveals if there’s a trend between price and the number of lectures.<h3>

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data,x="price",y="num_lectures",alpha=0.5)
plt.title("Course Price vs. Number of Lectures")
plt.show()
```

Interpretation:
Higher-priced courses tend to have more lectures, but there is no strict correlation.

Some low-cost courses also have many lectures, suggesting price is not the only factor affecting content quantity.

Most courses cluster below 200 lectures, but some outliers exceed 500+ lectures, especially at the higher price range (above $150).

Some expensive courses have fewer lectures, indicating that pricing may be based on factors like course quality, instructor reputation, or exclusivity rather than just lecture count.

# Trends & Growth Analysis


<h3>10. Course Publishing Trends Over the Years<br>
Question: Are online courses increasing in popularity over time?<br>
Why? To see if the number of courses is growing each year.<br>
Graph: Line Chart (df['Year'].value_counts().sort_index().plot(kind='line')) – Shows the trend of course publications over time.</h3>

```python
plt.figure(figsize=(graph_width, graph_height))
course_trend = data["Year"].value_counts().sort_index()
sns.lineplot(x=course_trend.index, y=course_trend.values, marker="o", color=udemy_purple)
for x, y in zip(course_trend.index, course_trend.values):
    plt.annotate(f"{y}", (x, y), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10, color=udemy_black)
plt.xlabel("Year", fontsize=12, color=udemy_black)
plt.ylabel("Number of Courses", fontsize=12, color=udemy_black)
plt.savefig("Course_creation_trend .png", dpi=300, bbox_inches='tight')
plt.show()
```

The number of courses published increased steadily from 2011 to 2016, peaking in 2016 i.e 2016 the highest number of published. However, there was a sharp decline in 2017, suggesting a possible saturation or change in demand.

<h3>13. Content Duration vs Subscribers<br>
Question: Do longer courses get more subscribers, or do students prefer shorter ones<br>
Why? To analyze if course length affects enrollment.<br>
Graph: Scatterplot (sns.scatterplot) – Shows the relationship between content duration and the number of subscribers.</h3>

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x='content_duration_minutes', y='num_subscribers', alpha=0.5)
plt.title("Content Duration vs. Subscribers")
plt.show()
```

<h3>14. Correlation Between Features<br>
Question: What are the key correlations between numerical features in the dataset?<br>
Why? To explore relationships between price, subscribers, reviews, duration, etc.<br>
Graph: Heatmap (sns.heatmap) – Provides an overview of correlations between numerical variables.</h3>

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=0)
plt.title("Correlation Matrix")
plt.show()
```

The number of subscribers is mostly driven by engagement factors (reviews, lectures) rather than price or duration.

The course length (content duration) and number of lectures are closely linked.

Older courses tend to have more subscribers due to their established presence.

```python
# 16 Forecasting - Future Growth in Revenue 
import matplotlib.ticker as mtick
plt.figure(figsize=(7,4))
revenue_trend = data.groupby("Year")["Revenue"].sum().sort_index()
sns.lineplot(x=revenue_trend.index, y=revenue_trend.values, marker="o", color=udemy_purple)
for x, y in zip(revenue_trend.index, revenue_trend.values):
    plt.text(x, y, f"{y/1e6:.1f}M", fontsize=10, ha='right', va='bottom', color=udemy_black)  # Display in Millions
plt.xlabel("Year", fontsize=12, color=udemy_black)
plt.ylabel("Total Revenue ($)", fontsize=12, color=udemy_black)
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))  # Y-axis in Millions
plt.savefig("Future_Growth_in_Revenue.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ⚠️ 5. Identifying Bottlenecks - Drop in Course Creation (Bar Chart)
plt.figure(figsize=(graph_width, graph_height))
ax=sns.barplot(x=course_trend.index, y=course_trend.values, color=udemy_purple)
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width() / 2,  # X position (center of the bar)
        p.get_height(),  # Y position (top of the bar)
        f'{int(p.get_height()):,}',  # Convert to integer format with commas
        ha='center', va='bottom', fontsize=10, color=udemy_black
    )
plt.xlabel("Year", fontsize=12, color=udemy_black)
plt.ylabel("Number of Courses", fontsize=12, color=udemy_black)
plt.xticks(color=udemy_black)
plt.savefig("Drop_in_Course_creation.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#7. Market Positioning - Subject-wise Revenue (Bar Chart)
plt.figure(figsize=(graph_width, graph_height))
subject_revenue = data.groupby("subject")["Revenue"].sum().sort_values(ascending=False)
ax=sns.barplot(x=subject_revenue.index, y=subject_revenue.values, color=udemy_purple)
for p in ax.patches:
    plt.text(
        p.get_x() + p.get_width() / 2,  # X position (center of bar)
        p.get_height(),  # Y position (top of bar)
        f'{p.get_height()/1e6:.1f}M',  # Convert to million format
        ha='center', va='bottom', fontsize=10, color=udemy_black
    )
plt.xlabel("Subject", fontsize=12, color=udemy_black)
plt.ylabel("Total Revenue ($)", fontsize=12, color=udemy_black)
plt.xticks(rotation=25, color=udemy_black)
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.savefig("Subject_wise_Revenue.png", dpi=300, bbox_inches='tight')
plt.show()
```
